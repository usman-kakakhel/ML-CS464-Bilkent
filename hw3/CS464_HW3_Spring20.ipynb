$I(T, A) = H(T) - \sum_{v=vals(A)}P_A(v)H(S_A(v))$

```python
from google.colab import drive
drive.mount('/content/gdrive')
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
import torch.nn.functional as F
from PIL import Image
import time
import os
import matplotlib.pyplot as plt

#the path where images folder and labels.xlsx is present. Also, i am saving my 
#model.pth and Convmodel.pth in this directory.
hw_path = "/content/gdrive/My Drive/Colab Notebooks/ML_HW3"

#average meter is used because in train method acc.avg etc was used and 
#internet suggested that this class is usually used to achieve averages of 
#losses and accuracies etc
class AverageMeter():
    def __init__(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val):
        self.val = val
        self.sum += val
        self.count += 1
        self.avg = self.sum / self.count

def plotting(all_losses_train, all_accuracies_train, all_losses_val, all_accuracies_val):
    x_axis = np.arange(len(all_losses_train))
    #training and validation losses in y axis and iteration in the x axis
    plt.plot(x_axis, all_losses_train, label='Training', lw=2)
    plt.plot(x_axis, all_losses_val, label='Validation', lw=2)
    plt.xlabel('The number of iterations')
    plt.ylabel('Loss')
    plt.title('Traning and Validation Loss')
    plt.legend()
    plt.show()

    #training and validation losses in y axis and iteration in the x axis
    plt.plot(x_axis, all_accuracies_train, label='Training', lw=2)
    plt.plot(x_axis, all_accuracies_val, label='Validation', lw=2)
    plt.xlabel('The number of iterations')
    plt.ylabel('Accuracy')
    plt.title('Traning and Validation Accuracy')
    plt.legend()
    plt.show()

def testingModel(path, test_loader):
    model = torch.load(path)
    model.eval()

    with torch.no_grad():
        for batch_idx, (data, labels) in enumerate(test_loader):
            predicted_labels = model.forward(data).squeeze()
            predictions = ((predicted_labels.cpu().detach().numpy() >= 0.5).astype(int)).flatten()
            print("Accuracy: ", np.mean(predictions == labels.cpu().detach().numpy()) * 100)
            TP = sum(predictions[np.where(predictions == labels.cpu().detach().numpy())] == 1)
            TN = sum(predictions[np.where(predictions == labels.cpu().detach().numpy())] == 0)
            FP = sum(predictions[np.where(predictions != labels.cpu().detach().numpy())] == 1)
            FN = sum(predictions[np.where(predictions != labels.cpu().detach().numpy())] == 0)
            precision = TP / (TP + FP)
            recall = TP / (TP + FN)
            F1 = 2 * precision * recall / (precision + recall)
            print("Precision: ", precision)
            print("Recall: ", recall)
            print("F1: ", F1)
            print("----Confusion Matrix----")
            print("TP: ", TP, "FP: ", FP)
            print("FN: ", FN, "TN: ", TN)
```

#### Data Loader [7pts]

An important part of such a task is to implement your own data loader. In this homework, a partial loader is provided to you. This loader is going to be based on a base class named "Dataset", provided in PyTorch library. You need to complete the code below to create your custom "OcularDataset" class which will be able to load your dataset. <br><br>
Implement the functions whose proptotypes are given. Follow the TODO notes below. You have to divide the files into three sets as <b>train (5/7)</b>, <b>validation (1/7)</b> and **test (1/7)** sets.  These non-overlapping splits, which are subsets of OcularDataset, should be retrieved using the "get_dataset" function. Here, you are also supposed to flatten the image into a vector (also to grayscale) to be compatible with MLP. Note that the pixel values also needs to be normalized to [0,1] range.
<br><br>

Hint: The dataset is not normalized and your results will heavily depend on your input.


```python
class OcularDataset(Dataset): 
    def __init__(self, dataSet):
        print("----------OcularDataset MLP----------")
        self.data = torch.FloatTensor(dataSet.astype('float')).cuda()
        
    '''This function should return sample count in the dataset'''
    def __len__(self):
        return self.data.shape[0]

    '''This function should return a single sample and its ground truth value from the dataset corresponding to index parameter '''
    def __getitem__(self, index):
        _x = self.data[index, :-1]
        _y = self.data[index, -1]
        return _x, _y

def get_dataset(root):
    print("----------Getting MLP dataset----------")
    label_file = pd.read_excel(root + '/labels.xlsx')
    label_file = label_file.to_numpy()
    X = np.zeros((label_file.shape[0], 32768))
    for i in range(label_file.shape[0]):
        im = Image.open(root + "/images/" + str(label_file[i,0]) +".jpg")
        im = np.array(im.convert("L")).flatten()
        X[i] = im
    #normalize
    X = (X - X.min())/(X.max() - X.min())
    dataSet = np.c_[X, label_file[:,1]]
    dataSet = dataSet[np.random.permutation(dataSet.shape[0]),:]
    cntTrain = int(label_file.shape[0] * (5/7))
    cntValidate = int(label_file.shape[0] * (1/7))
    cntTest =  int(label_file.shape[0] - (cntTrain + cntValidate))
    train_dataset = OcularDataset(dataSet[:cntTrain,:])
    val_dataset = OcularDataset(dataSet[cntTrain:cntTrain+cntValidate,:])
    test_dataset = OcularDataset(dataSet[-cntTest:,:])
    full_train_dataset = OcularDataset(dataSet[:cntTrain+cntValidate,:])

    return train_dataset, val_dataset, test_dataset, full_train_dataset
```

#### Neural Network [4 pts]

Now, implement your three hidden layer neural network. FNet class will represent your neural network. The layer descriptions are as follows:<ul>
    <i>> Input layer will have ReLU activation. You should decide the number of input neurons.</i><br>
    <i>> First hidden layer will have 1024 neuros with ReLU activation </i><br>
    <i>> Second hidden layer will have 256 neuros with ReLU activation </i><br>
    <i> You should decide the number of output neurons and pick a proper activation function for the output layer. </i><br>
</ul>

```python
class FNet(nn.Module):
    '''Define your neural network'''
    def __init__(self, inputSize=32768, outputSize=1): 
        super(FNet, self).__init__()
        self.inp = inputSize
        self.out = outputSize
        self.h1 = nn.Linear(self.inp, 1024)
        self.h2 = nn.Linear(1024, 256)
        self.output = nn.Linear(256, self.out)
      
        self.device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
        self.to(self.device)
    #sigmoid applied at the output layer
    def forward(self, X): 
        x = F.relu(X)
        x = F.relu(self.h1(x))
        x = F.relu(self.h2(x))
        return torch.sigmoid(self.output(x))
```

#### Training [10 pts]

Complete the code snippet below to train your network. You need to carefully select the appropriate loss function and tune hyper-parameters. Use SGD optimizer for this question.<br>

```python
train_batch = 25
test_batch = 500
best_path = hw_path
#got datasets from get_dataset methods
train_dataset, val_dataset, test_dataset, full_train_dataset = get_dataset(hw_path)
#make DataLoaders from all datasets
trainloader = DataLoader(train_dataset, batch_size=train_batch, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=test_batch)
test_loader = DataLoader(test_dataset, batch_size=test_batch)
full_train_loader = DataLoader(full_train_dataset, batch_size=train_batch, shuffle=True)
```

```python
#train batch is 25 as 25 is a factor of 2500
#test batch is 500 to complete testing in one go and have batches in testing
max_epoch = 100
#after much checking, this lr was selected
learning_rates = [ 0.01, 0.003, 0.001]



def main():
    best_loss = 100
    all_lr_losses_train = []
    all_lr_accuracies_train = []
    all_lr_losses_val = []
    all_lr_accuracies_val = []
    # select a learning rate
    for lr in learning_rates:
        #make model and optimizer and loss function
        model = FNet()
        optimizer = torch.optim.SGD(model.parameters(), lr=lr, momentum=0.9, weight_decay=5e-04)
        #criterion is BCELoss as cross entropy is usually used with sigmoid for classification
        criterion = nn.BCELoss()
        
        all_losses_train = []
        all_accuracies_train = []
        all_losses_val = []
        all_accuracies_val = []
        for epoch in range(max_epoch):
            losses, accuracies = train(epoch, model, criterion, optimizer, trainloader)
            all_losses_train.append(losses)
            all_accuracies_train.append(accuracies)

            loss, acc = test(model, val_loader, criterion)
            all_losses_val.append(loss)
            all_accuracies_val.append(acc)
            #using lowest loss instead of highest accuracy for saving model
            if loss < best_loss:
                best_loss = loss
                best_epoch = epoch
                best_lr = lr
                # best_model = model
        all_lr_losses_train.append(all_losses_train)
        all_lr_accuracies_train.append(all_accuracies_train)
        all_lr_losses_val.append(all_losses_val)
        all_lr_accuracies_val.append(all_accuracies_val)
    
    print("Best Learning: ", best_lr)
    # final training
    model = FNet()
    optimizer = torch.optim.SGD(model.parameters(), lr=best_lr, momentum=0.9, weight_decay=5e-04)
    criterion = nn.BCELoss()
    min_dif = 100
    for epoch in range(max_epoch):
        loss, acc = train(epoch, model, criterion, optimizer, full_train_loader)
        dif = loss - best_loss

        if (dif < min_dif):
            best_model = model
            min_diff = dif

    if os.path.exists(best_path + '/model.pth'):
        os.remove(best_path + '/model.pth')
    torch.save(best_model, best_path + '/model.pth')

    ind = learning_rates.index(best_lr)
    return all_lr_losses_train[ind], all_lr_accuracies_train[ind], all_lr_losses_val[ind], all_lr_accuracies_val[ind]
    
''' Train your network for a one epoch '''
def train(epoch, model, criterion, optimizer, trainloader):
    batch_time = AverageMeter()
    losses = AverageMeter()
    accuracies = AverageMeter()
    data_time = AverageMeter()
    model.train()
    datastart = time.time()

    for batch_idx, (data, labels) in enumerate(trainloader):
        data_time.update(time.time() - datastart)
        batchstart = time.time()
        #get labels from forward and then correct weights by backward propagation
        optimizer.zero_grad()
        predicted_labels = model.forward(data).squeeze()
        curr_loss = criterion(predicted_labels, labels)
        curr_loss.backward()
        optimizer.step()
        #accuracies and losses stored for current batch
        predictions = ((predicted_labels.cpu().detach().numpy() >= 0.5).astype(int)).flatten()
        accuracies.update(np.mean(predictions == labels.cpu().detach().numpy()) * 100)
        losses.update(curr_loss)
        batch_time.update(time.time() - batchstart)

        print('Epoch: [{0}][{1}/{2}]\t'
              'Time {batch_time.val:.3f} ({batch_time.avg:.3f})\t'
              'Data {data_time.val:.4f} ({data_time.avg:.4f})\t'
              'Loss {loss.val:.4f} ({loss.avg:.4f})\t'
              'Accu {acc.val:.4f} ({acc.avg:.4f})\t'.format(
                epoch + 1, batch_idx + 1, len(trainloader), 
                batch_time=batch_time,
                data_time=data_time, 
                loss=losses,
                acc=accuracies))
        datastart = time.time()

        
    return losses.avg , accuracies.avg

''' Test&Validate your network '''
def test(model, testloader, criterion):
    batch_time = AverageMeter()
    accuracies = AverageMeter()
    model.eval()

    with torch.no_grad():
        for batch_idx, (data, labels) in enumerate(testloader):
            batchstart = time.time()

            predicted_labels = model.forward(data).squeeze()
            curr_loss = criterion(predicted_labels, labels)

            predictions = ((predicted_labels.cpu().detach().numpy() >= 0.5).astype(int)).flatten()
            batch_time.update(time.time() - batchstart)
            accuracies.update(np.mean(predictions == labels.cpu().detach().numpy()) * 100)

            print('Time {batch_time.avg:.3f}\t'
                'Accu {acc.avg:.4f}\t'.format(
                batch_time=batch_time, 
                acc=accuracies))
    
    return curr_loss.item(), accuracies.val

all_losses_train, all_accuracies_train, all_losses_val, all_accuracies_val = main()
```

#### Data Loader [5 pts]

In this part, you will train a CNN for the same problem. Again, the pixel values need to be normalized to [0,1] range. Please do **not** change images to grayscale this time. First, implement the data loader (OcularDataset). You have to divide the files into three sets which are <b>train (5/7)</b>, <b>validation (1/7)</b> and **test (1/7)**.  These non-overlapping splits, which are subsets of OcularDataset, should be retrieved using the "get_dataset" function.<br> You may use your data loader from the previous sections with propoer modifications. Note that this time you do **not** need to flatten the image. 

```python
class OcularDataset(Dataset):
    def __init__(self, X, Y):
        print("----------OcularDataset CNN----------")
        self.dataX = torch.FloatTensor(X.astype('float')).cuda()
        self.dataY = torch.FloatTensor(Y.astype('float')).cuda()
    
    def __len__(self):
        return len(self.dataY)
    
    def __getitem__(self, index):
        _x = self.dataX[index]
        _y = self.dataY[index]
        return _x, _y
        
def get_dataset(root):
    print("----------Getting CNN dataset----------")
    label_file = pd.read_excel(root + '/labels.xlsx')
    label_file = label_file.to_numpy()
    X = np.zeros((label_file.shape[0], 3, 128, 256))
    for i in range(label_file.shape[0]):
        im = Image.open(root + "/images/" + str(label_file[i,0]) +".jpg")
        im = np.array(im)
        X[i] = np.stack((im[:,:,0], im[:,:,1], im[:,:,2]), axis=0)
    #normalize
    X = (X - X.min())/(X.max() - X.min())
    #shuffeling data set before getting test train and validation set
    shuffled_indexes = np.random.permutation(len(label_file[:,1]))
    X = X[shuffled_indexes,:,:]
    Y = label_file[shuffled_indexes,1]
    cntTrain = int(label_file.shape[0] * (5/7))
    cntValidate = int(label_file.shape[0] * (1/7))
    cntTest =  int(label_file.shape[0] - (cntTrain + cntValidate))
    train_dataset = OcularDataset(X[:cntTrain], Y[:cntTrain])
    val_dataset = OcularDataset(X[cntTrain:cntTrain+cntValidate], Y[cntTrain:cntTrain+cntValidate])
    test_dataset = OcularDataset(X[-cntTest:], Y[-cntTest:])
    full_train_dataset = OcularDataset(X[:cntTrain+cntValidate], Y[:cntTrain+cntValidate])

    return train_dataset, val_dataset, test_dataset, full_train_dataset
```

#### Convolutional Neural Network [8 pts]

Now implement your CNN. ConvNet class will represent your convolutional neural network. Implement 3 layers of convolution: 
1. <i>> 8 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br> 
    <i>> ReLU </i><br>
2. <i>> 16 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br>
    <i>> ReLU </i><br>
    <i>> MaxPool 2 x 2 </i><br>   
3. <i>> 32 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br>
    <i>> ReLU </i><br>
    <i>> MaxPool 2 x 2 </i><br>

As a classification layer, you need to add only one more fully-connected layer at the end of the network. You need to choose the appropriate input and output neuron sizes and the activation function.

```python
class ConvNet(nn.Module):
    '''Define your neural network'''
    def __init__(self): 
        super(ConvNet, self).__init__()
        
        self.h1 = nn.Conv2d(3, 8, 3, 1)
        self.h2 = nn.Conv2d(8, 16, 3, 1)
        self.pool1 = nn.MaxPool2d(2)
        self.h3 = nn.Conv2d(16, 32, 3, 1)
        self.pool2 = nn.MaxPool2d(2)
        self.output = nn.Linear(59520, 1)

        self.device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
        self.to(self.device)
    #sigmoid applied at the output layer
    def forward(self, X):
        x = F.relu(self.h1(X))
        x = F.relu(self.h2(x))
        x = self.pool1(x)
        x = F.relu(self.h3(x))
        x = self.pool2(x)
        x = x.view(x.size()[0], -1)
        return torch.sigmoid(self.output(x))
```

```python
#train batch is 25 as 25 is a factor of 2500
#test batch is 500 to complete testing in one go and have batches in testing
max_epoch = 150
#after much checking, this lr was selected
learning_rate = 0.001

def main():
    best_loss = 100
    #make model and optimizer and loss function
    model = ConvNet()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9, weight_decay=5e-04)
    #criterion is BCELoss as cross entropy is usually used with sigmoid for classification
    criterion = nn.BCELoss()
    
    all_losses_train = []
    all_accuracies_train = []
    all_losses_val = []
    all_accuracies_val = []
    for epoch in range(max_epoch):
        losses, accuracies = train(epoch, model, criterion, optimizer, trainloader)
        all_losses_train.append(losses)
        all_accuracies_train.append(accuracies)

        loss, acc = test(model, val_loader, criterion)
        all_losses_val.append(loss)
        all_accuracies_val.append(acc)
        #using lowest loss instead of highest accuracy for saving model
        if (loss < best_loss):
            best_loss = loss
            # best_model = model
    
    # final training
    model = ConvNet()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9, weight_decay=5e-04)
    criterion = nn.BCELoss()
    min_dif = 100
    for epoch in range(max_epoch):
        loss, acc = train(epoch, model, criterion, optimizer, full_train_loader)
        dif = loss - best_loss

        if (dif < min_dif):
            best_model = model
            min_diff = dif

    if os.path.exists(best_path + '/Convmodel.pth'):
        os.remove(best_path + '/Convmodel.pth')
    torch.save(best_model, best_path + '/Convmodel.pth')

    return all_losses_train, all_accuracies_train, all_losses_val, all_accuracies_val
    
''' Train your network for a one epoch '''
def train(epoch, model, criterion, optimizer, trainloader):
    batch_time = AverageMeter()
    losses = AverageMeter()
    accuracies = AverageMeter()
    data_time = AverageMeter()
    model.train()
    datastart = time.time()

    for batch_idx, (data, labels) in enumerate(trainloader):
        data_time.update(time.time() - datastart)
        batchstart = time.time()
        #get labels from forward and then correct weights by backward propagation
        optimizer.zero_grad()
        predicted_labels = model.forward(data).squeeze()
        curr_loss = criterion(predicted_labels, labels)
        curr_loss.backward()
        optimizer.step()
        #accuracies and losses stored for current batch
        predictions = ((predicted_labels.cpu().detach().numpy() >= 0.5).astype(int)).flatten()
        accuracies.update(np.mean(predictions == labels.cpu().detach().numpy()) * 100)
        losses.update(curr_loss)
        batch_time.update(time.time() - batchstart)

        print('Epoch: [{0}][{1}/{2}]\t'
              'Time {batch_time.val:.3f} ({batch_time.avg:.3f})\t'
              'Data {data_time.val:.4f} ({data_time.avg:.4f})\t'
              'Loss {loss.val:.4f} ({loss.avg:.4f})\t'
              'Accu {acc.val:.4f} ({acc.avg:.4f})\t'.format(
                epoch + 1, batch_idx + 1, len(trainloader), 
                batch_time=batch_time,
                data_time=data_time, 
                loss=losses,
                acc=accuracies))
        datastart = time.time()

        
    return losses.avg , accuracies.avg

''' Test&Validate your network '''
def test(model, testloader, criterion):
    batch_time = AverageMeter()
    accuracies = AverageMeter()
    model.eval()

    with torch.no_grad():
        for batch_idx, (data, labels) in enumerate(testloader):
            batchstart = time.time()

            predicted_labels = model.forward(data).squeeze()
            curr_loss = criterion(predicted_labels, labels)

            predictions = ((predicted_labels.cpu().detach().numpy() >= 0.5).astype(int)).flatten()
            batch_time.update(time.time() - batchstart)
            accuracies.update(np.mean(predictions == labels.cpu().detach().numpy()) * 100)

            print('Time {batch_time.avg:.3f}\t'
                'Accu {acc.avg:.4f}\t'.format(
                batch_time=batch_time, 
                acc=accuracies))
    
    return curr_loss.item(), accuracies.val

all_losses_train, all_accuracies_train, all_losses_val, all_accuracies_val = main()


```
